```python
import os
import numpy as np
import math

###############################
#DISTANCE COMPUTATION
###############################



#Def pictures is used to compute distance between to places of SF_XS Database
#using the pictures name to compute distances with GEO PY Lib
#Ground truth means which reference are positive for each query
def picturesDist(img1,img2):

#Def pictures is used to compute distance between to places of SF_XS Database
#using the pictures name to compute distances with GEO PY Lib
#Ground truth means which reference are positive for each query

  img1split = img1.split('@')
  img1coord1  = float(img1split[1])
  img1coord2 = float(img1split[2])

  img2split = img2.split('@')
  img2coord1 = float(img2split[1])
  img2coord2 = float(img2split[2])

  # Compute distance between two points
  dist = math.sqrt((img1coord1 - img2coord1)**2 + (img1coord2 - img2coord2)**2)

  return dist

```

```python
###############################
#SAN FRANCISCO XS TEST AND VAL
###############################

#TRESHOLD IN METERS: represents the max distance between two GPS coordinates
#to identify the same place
THRESHOLD = 25

# Specify folder path for San Francisco VAL
ROOT_PATH_SF_VAL = 'val'
ROOT_PATH_SF_TEST = 'test'
ROOTS = []
ROOTS.append(ROOT_PATH_SF_VAL)
ROOTS.append(ROOT_PATH_SF_TEST)



#generate .npy
for paths in ROOTS:

  ROOT = '/content/drive/MyDrive/Datasets/sf_xs/'
  ROOT_PATH =os.path.join(ROOT,paths)

  # reference images names
  #sfval_dbImages.npy
  database = 'database'
  folder_path_reference = os.path.join(ROOT_PATH,database)
  dbImages_list = os.listdir(folder_path_reference)
  dbImages_array = np.array(dbImages_list)
  dbImages_final = np.core.defchararray.add('database/', dbImages_array)



  #query images name
  #sfval_qImages.npy
  query = 'queries'
  folder_path_query = os.path.join(ROOT_PATH,query)
  qImages_list = os.listdir(folder_path_query)
  qImages_array = np.array(qImages_list)
  qImages_final = np.core.defchararray.add('queries/', qImages_array)


  #ground truth
  #sfval_gt.npy

  #create ground truth list of list
  gt_list = []
  #Iterate on all query images
  for file_query in qImages_list:
    #create a list for each picture in the queries database
    gt_element = []
    #Read the name of query pictures
    file_query_path = os.path.join(folder_path_query,file_query)
    #Iterate on all reference pictures
    counter = 0
    for file_reference in dbImages_list:

      #Read the name of reference pictures
      file_reference_path = os.path.join(folder_path_reference,file_reference)
      #Compute distance between reference and query using the coordinates in the
      #name
      dist = picturesDist(file_query_path,file_reference_path)
      #If distance less then or equal to threshold
      if dist <= THRESHOLD:
        #Add the reference pictures name to the query db images linked
        gt_element.append(counter)
      counter = counter+1

    #Add gt_element to gt
    gt_list.append(np.array(gt_element))

  gt_final = np.array(np.array(gt_list,dtype=object))

  DBIMAGESNPY = os.path.join(ROOT,paths+'_dbImages.npy')
  QIMAGESNPY = os.path.join(ROOT,paths+'_qImages.npy')
  GTIMAGESNPY = os.path.join(ROOT,paths+'_gt.npy')

  np.save(DBIMAGESNPY, dbImages_final)
  np.save(QIMAGESNPY, qImages_final)
  np.save(GTIMAGESNPY, gt_final)


```
